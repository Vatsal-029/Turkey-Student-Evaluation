## Clustering

```python
dataset_questions = df.iloc[:,5:33]
```

```python
dataset_questions.head()
```

```python
# PCA for feature dimensional reduction
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
dataset_questions_pca = pca.fit_transform(dataset_questions)
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wsse = [] #within the clusters some of squared errors & inertia calculates those errors
for i in range(1, 11): #11 is excluding
    kmeans = KMeans(n_clusters = i, random_state = 10) #random_state is equivalent to set.seed, but number should be same so that models are matched
    kmeans.fit(dataset_questions_pca)
    wsse.append(kmeans.inertia_) # append helps to add a single element at a time
plt.plot(range(1, 11), wsse)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()

#from the graph we find that 5 is the optimum value of k, from where the errors start declining gradually.
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(dataset_questions_pca)
```

```python
y_kmeans
```

```python
# Visualising the clusters
plt.scatter(dataset_questions_pca[y_kmeans == 0, 0], dataset_questions_pca[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Negative Feedback')
plt.scatter(dataset_questions_pca[y_kmeans == 1, 0], dataset_questions_pca[y_kmeans == 1, 1], 
            s = 100, c = 'orange', label = 'Neutral Feedback')
plt.scatter(dataset_questions_pca[y_kmeans == 2, 0], dataset_questions_pca[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Positive Feedback')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'blue', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
